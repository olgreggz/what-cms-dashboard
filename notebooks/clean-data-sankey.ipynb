```python
df = (
    pd.read_csv(RAW_DATA_FILEPATH, encoding="ISO-8859-1")
    .loc[
        lambda df: ~df["Industry"].isin(EXCLUDE_INDUSTRIES),
        COLUMNS + [REVENUE_COLUMN] + ["Industry"],
    ]
    .rename(columns={REVENUE_COLUMN: "revenue"})
    .assign(
        cms_2020=lambda df: df["WHOCMS Previous 2020"].apply(group_cms),
        cms_2021=lambda df: df["WHOCMS Previous 2021"].apply(group_cms),
        cms_now=lambda df: df["WHOCMS"].apply(group_cms),
        count=True,
        revenue=lambda df: df["revenue"].str.replace(",", "").astype("float"),
    )
    .drop(COLUMNS + ["Industry"], axis=1)
    .dropna()
)
```

```python
agg_df = (
    df.groupby(["cms_2021", "cms_now"])["count", "revenue"]
    .agg({"revenue": "sum", "count": "count"})
    .reset_index()
    .loc[lambda df: df["cms_2021"] != df["cms_now"]]
    .loc[lambda df: (df["cms_2021"] == "Finalsite") | (df["cms_now"] == "Finalsite")]
)
```

```python
# note: the data that is labeled as 2021 is actually 2020, I'm renaming it here
plot_df = agg_df.assign(
    cms_2020=lambda df: df["cms_2021"] + " 2020",
    cms_now=lambda df: df["cms_now"] + " Current",
)


labels = list(set(plot_df["cms_2020"].values) | set(plot_df["cms_now"].values))
label_to_idx_lookup = {e: i for i, e in enumerate(labels)}

plot_df = plot_df.assign(
    cms_2020_idx=lambda df: df["cms_2020"].replace(label_to_idx_lookup),
    cms_now_idx=lambda df: df["cms_now"].replace(label_to_idx_lookup),
)
```
